```python
import pandas as pd
df = pd.read_csv('Advertising.csv')
df.head()
```

```python
X = df.drop(['sales','radio','newspaper'],axis=1)
y = df['sales'].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
```

```python
import seaborn as sns
sns.pairplot(df,x_vars=['TV'],y_vars=['sales'],kind='reg',height=7,aspect=0.7)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
rmse = (mse)**(1/2)
print("Mean Squared Error",mse)
print("Root Mean Squared Error",rmse)
print("Mean Absolute Error",mae)
```

```python
X = df.drop(['sales','TV','newspaper'],axis=1)
y = df['sales'].values.reshape(-1,1)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
reg=LinearRegression()
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
sns.pairplot(df,x_vars=['radio'],y_vars=['sales'],kind='reg',height=7,aspect=0.7)
```
